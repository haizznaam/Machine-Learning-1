# Problem 1: SVM Aglorithm Transformation

The two-class classification problem using linear models of the form
$$
y(x)=w^T \phi(x)+b
$$
where $\phi(x)$ denotes a fixed X-space transformation.

The training data set comprises N input vectors x1, ..., xN , with
corresponding target values t1, ...,tN where tn ∈ {−1, 1}, and new data
points x are classified according to the sign of y(x).

We shall assume that the training data set is linearly separable in X
space 

⇒ there exists at least one choice of the parameters w and b
satisfies y(xn) > 0 for points having tn = +1 and y(xn) < 0 for points
having tn = −1, so that tny(xn) > 0 for all training data points.

The distance of a point xn to the decision surface defined by y(x) = 0 is
given by:
$$d(x_n, y(x)) = \frac{|b+\mathbf{w}^{\intercal}\mathbf{\phi(x)}_i|}{||\mathbf{w}||_2} = \frac{t_n(b+\mathbf{w}^{\intercal}\mathbf{\phi(x)}_n)}{||\mathbf{w}||_2}$$

The margin is given by the perpendicular distance to the closest point xn
from the data set, and we wish to optimize the parameters w and b in
order to maximize this distance. Thus the maximum margin solution is
found by solving:
$$\hat{\mathbf{w}}, \hat{b}=\arg \max \left\{\min _{n} \frac{t_n\left(\mathbf{w}^{\top} \mathbf{\phi(x_n)+b}\right)}{\|\mathbf{w}\|_2}\right\}$$

We note that if we make the rescaling $w → αw$ and $b → αy$, then the
distance from any point $x_n$ to the decision surface, given by $\frac{t_n y(x_n)}{||\mathbf{w}||}$ is unchanged. We can use this freedom to set:
$$t_n(\mathbf{w}^{\intercal} {\phi(x_n)}  + b) = 1$$

for the point that is closest to the surface


**Optimization problem**: 

$$\begin{aligned} 
\hat{\mathbf{w}}, \hat{b} & =\arg \max \frac{1}{\|\mathbf{w}\|^2_2} \\ 
\text { subject } & :t_n(\mathbf{w}^{\intercal} {\phi(x_n)}  + b) \geq 1, \forall n=\overline{1, N}
\end{aligned}$$

**To simplify, the optimization problem can invert the objective function to convert to equivalent form:**
$$\begin{aligned} 
\hat{\mathbf{w}}, \hat{b} & =\arg \min \|\mathbf{w}\|^2_2 \\ 
\text { subject } & :t_n(\mathbf{w}^{\intercal} {\phi(x_n)}  + b) \geq 1, \forall n=\overline{1, N}
\end{aligned}$$

In order to solve this constrained optimization problem, we introduce
Lagrange multipliers $a_n ≥ 0$, with one multiplier an for each of the
constraints, giving the **Lagrangian function**:
$$\mathcal{L}(\mathbf{w},\mathbf{b},\mathbf{a}) = \min_{\mathbf{w}, b, \lambda} \mathcal{L}(\mathbf{w}, b, \lambda) = \frac{1}{2} ||\mathbf{w}||_2^2 - \sum_{n=1}^N a_n(t_n(\mathbf{w}^{\intercal} {\phi(x_n)}  + b) -1) \\
$$

Setting the derivatives of L(w, b, a) with respect to w and b equal to
zero

$$\begin{aligned} 
& \frac{\partial \mathcal{L}(\mathbf{w},\mathbf{b},\mathbf{a})}{\partial \mathbf{w}}=\mathbf{w}-\sum_{n=1}^N a_n t_n \phi(\mathbf{x}_n)=0 \Rightarrow \mathbf{w}=\sum_{n=1}^N a_n t_n \phi(\mathbf{x}_n)\\ 
& \frac{\partial \mathcal{L}(\mathbf{w},\mathbf{b},\mathbf{a})}{\partial b}=-\sum_{n=1}^N a_n t_n=0 \Rightarrow \sum_{n=1}^N a_n t_n=0
\end{aligned}$$

Eliminating w and b from L(w, b, a) using these conditions then gives
the dual representation of the maximum margin problem in which we
maximize

$$
\tilde{L}(a)=\sum_{n=1}^N a_n-\frac{1}{2} \sum_{n=1}^N \sum_{m=1}^N a_n a_m t_n t_m k\left(x_n, x_m\right)
$$
$\text { subject to the constraints }:$
$$
\begin{aligned}
& a_n \geq 0 \\
& \sum_{n=1}^N a_n t_n=0
\end{aligned}
$$

Define the kernel function : $k\left(x,x^{'}\right) = \phi(x)^{\intercal} \phi(x^{'})$

Knowing the $a_i$ we can find the weights $\mathrm{w}$ for the maximal margin separating hyperplane.
$$
\begin{aligned}
& w=\sum_{n=1}^N a_n t_n \phi\left(x_n\right) \\
& y(x)=\sum_{n=1}^N a_n t_n k\left(x, x_n\right)+b .
\end{aligned}
$$

Kuhn-Tucker theorem: the solution we find here will be the same as the solution to the original problem. Also KKT conditions require that the following three properties hold:
$$
\begin{aligned}
a_n & \geq 0 \\
t_n y\left(x_n\right)-1 & \geq 0 \\
a_n\left(t_n y\left(x_n\right)-1\right) & =0
\end{aligned}
$$

Thus for every data point, either $a_n=0$ or $t_n y\left(x_n\right)=1$.
- **Any data point for which $a_n=0$ will not appear in the sum and hence plays no role in making predictions for new data points**

- **The remaining data points are called support vectors. Because they satisfy $t_n y\left(x_n\right)=1$, they correspond to points that lie on the maximum margin hyperplanes in X space**


We note that $t_n y\left(x_n\right)=1$ then:
$$
\begin{array}{r}
t_n\left(\sum_{m \in S} a_m t_m k\left(x_n, x_m\right)+b\right)=1 \\
\\
\Rightarrow b=\frac{1}{N_S} \sum_{n \in S}\left(t_n-\sum_{m \in S} a_m t_m k\left(x_n, x_m\right)\right)
\end{array}
$$

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv(r'D:\NEU\Năm 3\ML\GitHub ML\Machine-Learning-1\SVM\data\dataset1.csv')
df = df.drop(columns='Unnamed: 0')
df
```

```python
# split train data
y_train = df['y']
X_train = df.drop(columns='y')
```

**Visualize data ban đầu**

```python
# Visualize data ban đầu
def plot_data(X,y):
   neg = np.where(y==0)[0]
   pos = np.where(y==1)[0]

   plt.scatter(X.iloc[neg,0], X.iloc[neg,1],marker='o',label = 'negative',color = 'midnightblue')
   plt.scatter(X.iloc[pos,0], X.iloc[pos, 1], marker='+', label='positive', color='y')
   plt.xlabel('X1')
   plt.ylabel('X2')
   plt.legend()
   plt.show()

plot_data(X_train, y_train)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC

# trainning model
svm_model = SVC(kernel='rbf',probability=True)
svm_model.fit(X_train, y_train)

# make predictions
y_train_pred = svm_model.predict(X_train)
print(f'Accuracy of TRAIN data: {100*accuracy_score(y_train, y_train_pred)}')
```

**Visualize kết quả dự đoán (hint: có thể plot bộ dataset với color range là prediction probability)**

```python
xx, yy = np.meshgrid(np.linspace(X_train['X1'].min(), X_train['X1'].max(), 100),
                     np.linspace(X_train['X2'].min(), X_train['X2'].max(), 100))

probas = svm_model.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:, 1].reshape(xx.shape)

plt.contourf(xx, yy, probas, cmap='viridis',
             levels=np.linspace(0, 1, 10), alpha=0.8)
plt.colorbar()
plt.scatter(X_train['X1'], X_train['X2'], c=y_train_pred,
            cmap='viridis', edgecolors='k', marker='o')
plt.title('Decision boundary and test set')
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
```

**Optional: Tune thử các hyperpameter của SVM xem có thể đạt accuracy > 90% không**

```python
# - Optional: Tune thử các hyperpameter của SVM xem có thể đạt accuracy > 90% không
C = 100
gamma = 10

svm_model2 = SVC(C=C,
                 kernel='rbf',
                 gamma=gamma)

svm_model2.fit(X_train, y_train)

y_train_pred2 = svm_model2.predict(X_train)

print(f'Accuracy of TRAIN data: {100*accuracy_score(y_train, y_train_pred2)}')
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline

parameters = {
    'clf__kernel': ['linear', 'rbf', 'sigmoid'],  # Các dạng hàm kernel
    'clf__C': [5,10,100],  # Trọng số của phạt phân loại sai
    'clf__gamma': [2,4,10],  # Tương ứng với tham số gamma của đa thức
}


pipeline = Pipeline(
    steps=[("clf", SVC())]
)

gscv = GridSearchCV(pipeline, parameters,cv=20,n_jobs=20,
                    scoring='accuracy', return_train_score=True, verbose=3)
gscv.fit(X_train, y_train)
```

```python
print("Best Parameters:", gscv.best_params_)
print("Best Accuracy Score:", gscv.best_score_)
```

# Problem 3:

```python
data = pd.read_csv(r'D:\NEU\Năm 3\ML\GitHub ML\Machine-Learning-1\SVM\data\heart.csv')
data
```

**Split train test data**

```python
from sklearn.model_selection import train_test_split

y_full = data['target']
X_full = data.drop(columns='target')

X_train, X_test, y_train, y_test = train_test_split(X_full, y_full, train_size=0.8, test_size=0.2, random_state=0)
```

**Model evaluation**

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report, f1_score,  mean_squared_error, roc_auc_score

def evaluation(model):

    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    print("="*50)
    
    print(confusion_matrix(y_train, y_train_pred))
    print(classification_report(y_train, y_train_pred))
    print("Accuracy of TRAIN data:", 100*accuracy_score(y_train, y_train_pred))
    print("F1_Score of TRAIN data:", 100*f1_score(y_train, y_train_pred, average="macro"))
    print("="*50)
    
    print(confusion_matrix(y_test, y_test_pred))
    print(classification_report(y_test, y_test_pred))
    print("Accuracy of TEST data:", 100*accuracy_score(y_test, y_test_pred))
    print("F1_Score of TEST data:", 100*f1_score(y_test, y_test_pred, average="macro"))
    print("="*50)
    
    # calculate the RMSE
    rmse = np.sqrt(mean_squared_error(y_test, y_test_pred))
    print("RMSE: ", rmse)
    
    # calculate ROC AUC
    roc_auc = roc_auc_score(y_test, y_test_pred, average=None)
    print("ROC AUC score: ", roc_auc)
    
```

**Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(X_train,y_train)
evaluation(log_reg)
```

**Naive Bayes Classifier**

```python
from sklearn.naive_bayes import GaussianNB
GNB = GaussianNB()
GNB.fit(X_train,y_train)
evaluation(GNB)
```

**SVM**

```python
from sklearn.svm import SVC
svm = SVC()
svm.fit(X_train,y_train)
evaluation(svm)
```
